```python
import requests
from bs4 import BeautifulSoup

# 웹 페이지 요청
url = 'https://news.naver.com/main/list.naver?mode=LPOD&mid=sec&oid=032'
response = requests.get(url)

# HTML 파싱
soup = BeautifulSoup(response.text, 'html.parser')

ul_tags=soup.select('#main_content div.list_body.newsflash_body ul')
all_datas=[]

for ul in ul_tags:
    li_tags=ul.select('li')
    for li in li_tags:

        try:#사진을 클릭해도 되고 제목을 클릭해도 되기 때문에 둘 중 하나의 링크만 가져오기
            url=li.select('a')[0].attrs['href'] #사진 해당 링크
        except:
            url=li.select('a')[1].attrs['href'] #제목 해당 링크

        res2=requests.get(url)
        soup2=BeautifulSoup(res2.text,'html.parser')

        title=soup2.select_one('#title_area span').text.strip()
        body=soup2.select_one('#dic_area').find_all(string=True,recursive=False)# 왜 text가 안되죠???? 
        body=' '.join(body).strip()

        all_datas.append(dict(title=title,body=body))
print(len(all_datas))
print(all_datas[:2])

        
```

```python
import requests 
from bs4 import BeautifulSoup 

response = requests.get('https://crawlingstudy-dd3c9.web.app/01/')
soup = BeautifulSoup(response.text, 'html.parser')

a_tags = soup.select_one('#cook').text.strip()
print(a_tags)
```

```python
#----------------------------------key값 리스트에 담기 -------------------------------
th_tags = soup.select('thead th') 
print(th_tags) # 확인용 

# for문으로 key값 갖고 오기 
keys = []
for th in th_tags:
    keys.append(th.text.strip())
#print('keys:', keys) # ['이름', '나이']
print()

#----------------------------------value 값 리스트에 담기 -------------------------------

tr_tags = soup.select('tbody tr') # tr행 먼저 전부 갖고 오기 
#print('tr_tags:', tr_tags)
print()

# for문으로 data 갖고오기 
result = []
for tr in tr_tags: # tr 
    #print('tr: ', tr)
    print()
    values = [] # 두 번째 돌 때는 초기화 돼야함 (홍길동, 23 넣기 위해서)

    for td in tr.find_all('td'):
        # print('td: ', td.text)
        # print()
        values.append(td.text) # ['이몽룡', '34']
    
    result.append(dict(zip(keys, values))) # zip : 같은 인덱스 끼리 묶어줌 -> ex) {이름: 이몽룡, 나이: 34} 

result
```

```python
import json 
import requests
from bs4 import BeautifulSoup 


all_datas = []
page = 1
while True:

    naverNews_src = requests.get(f'https://news.naver.com/main/list.naver?mode=LPOD&mid=sec&oid=032&date=20240807&page={page}')
    soup = BeautifulSoup(naverNews_src.text, 'html.parser')

    if str(page) != soup.select_one('div.paging strong').text:
        print('크롤링 완료')
        break

    print('현재 페이지:', page)

    ul_tags = soup.select('div.list_body.newsflash_body ul')

    for ul_tag in ul_tags:

        li_tags = ul_tag.select('li')

        for li_tag in li_tags:

            try:
                url = li_tag.select('a')[0].attrs['href']
            except:
                url = li_tag.select('a')[1].attrs['href']

            news_res = requests.get(url)
            news_soup = BeautifulSoup(news_res.text, 'html.parser')
            

            title = '타이틀 부재'
            body = 'body 부재'

            if news_soup.select_one('#title_area span') != None:
                title = news_soup.select_one('#title_area span').text

            
            if news_soup.select_one('#dic_area') != None:
                # body = news_soup.select_one('#dic_area').text.split('\n')
                body = ''.join(news_soup.select_one('#dic_area').text.split('\n')).strip()
            

            all_datas.append(dict(
                page = page,
                title = title,
                body = body
            ))

    page += 1



print(len(all_datas))
import json 


with open('json_news.json', 'w', encoding='utf-8') as json_file:
    json.dump(all_datas, json_file, ensure_ascii=False)
```

```python
import requests 
import json 

res1 = requests.get('https://jsonplaceholder.typicode.com/photos')
json1 = res1.json()

result = []

for data in json1:
    # print(data)

    id = data['id']
    title = data['title']

    temp = dict(
        id = id,
        title = title,
        comments = []
    )

    res2 = requests.get(f'https://jsonplaceholder.typicode.com/comments?postId={id}')

    for comment in res2.json():
        temp['comments'].append(comment['body'])


    result.append(temp)

    if id == 10:
        break

with open('test2.json', 'w') as json_file:
    json.dump(result, json_file)
```

```python
import requests 
import json 

res1 = requests.get('https://display.musinsa.com/display/api/v2/categories/ITEM/goods?siteKindId=musinsa&sex=A&sortCode=pop_category&categoryCode=001&size=60&page=1')
json1 = res1.json()

result = []
print(json1)

for data in json1['data']['goodsList']:
    print(data)
    brandName = data['brandName']
    goodsName = data['goodsName']
    normalPrice=data['normalPrice']
    price=data['price']
    result.append(dict(
        brandName = brandName,
        goodsName = goodsName,
        normalPrice = normalPrice,
        price=price))
    

print(len(result))
with open('test3.json', 'w') as json_file:
    json.dump(result, json_file)
```
