```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
from selenium.webdriver import ActionChains #해당 스크롤 위치까지만 끌어올 수 있음. 해당 소스를 이용해서
from bs4 import BeautifulSoup
import json

driver=webdriver.Chrome()
driver.implicitly_wait(1)

driver.get('https://crawlingstudy-dd3c9.web.app/05/')
time.sleep(1)

div_tags=driver.find_elements(By.CSS_SELECTOR,'#post div[onclick]') #클릭할 수 있는 div 태그만 모아서 List로 뽑기
print(div_tags)

all_data=[]

old_height=driver.execute_script('return document.body.scrollHeight')#자바 활용해서 해당 페이지의 스크롤 길이 찾기

for idx,tag in enumerate(div_tags,1):#1부터 시작
        driver.execute_script('window.scrollTo(0,document.body.scrollHeight);')#함수로 스크롤 길이만큼 이동

        if idx==5:#5개만 뽑기 위해서
            break

        print(idx+1,tag.text) #title 뽑기

        tag.click()
        time.sleep(2)

        #타이틀 안에 여러개의 comments를 list return
        comments=driver.find_elements(By.CSS_SELECTOR,f'#\\3{idx}>div') 
##\31  > div:nth-child(2) 해당 코멘트의 selector가 이거임. 이걸 각 id의 div5개를 뽑아야함

        data=dict(
              id=idx+1,
              title=tag.text
        )

        #이메일을 제외하고 짝수에 해당하는 코멘트만 불러오기
        data['comments']=[comments[i].text for i in range(1,len(comments),2)]

        all_data.append(data)
        #break

driver.quit()


# with open['testDatas.json','w'] as json_file:
#       json.dump(all_data,json_file)
```
